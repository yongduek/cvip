```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
import imageio
```

```python
url = 'https://imgc.artprintimages.com/img/print/weeki-wachee-spring-florida_u-l-q10wjnw0.jpg'
im = imageio.imread(url)
im.shape, im.dtype
```

```python
plt.imshow (im)
```

```python
gamma_values = [3, 1.5, 1/2., 1/3., 1./4]
imlist = []
for gamma in gamma_values:
    imlist.append(np.clip(np.power (im/255., gamma) * 255., 0, 255).astype(np.uint8))
```

```python
plt.imshow (imlist[0])
plt.title ('gamma = {}'.format(gamma_values[0]))
```

```python
plt.imshow (imlist[-1])
plt.title ('gamma = {}'.format(gamma_values[-1]))
```

```python
imlist[0].shape
```

```python
plt.figure(figsize=(13,4))

plt.subplot(1,3,1)
plt.hist(imlist[0].ravel(), bins=255, density=True)
plt.title ('gamma {}'.format(gamma_values[0]))

plt.subplot(1,3,2)
plt.hist(im.ravel(), bins=255, density=True)

plt.subplot(1,3,3)
plt.hist(imlist[-1].ravel(), bins=255, density=True)
plt.title ('gamma {}'.format(gamma_values[-1]))
```

```python
x = np.linspace(0,1,255)
```

```python
plt.plot (x, np.power(x, gamma_values[0]))
plt.title ('gamma {}'.format(gamma_values[0]))
```

```python
plt.plot (x, np.power(x, gamma_values[-1]))
plt.title ('gamma {}'.format(gamma_values[-1]))
```
