# Image Segmentation, Watershed, CCL
- https://docs.opencv.org/master/d3/db4/tutorial_py_watershed.html
- https://docs.opencv.org/4.1.2/d2/dbd/tutorial_distance_transform.html
- https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_watershed/py_watershed.html
- `cv2.watershed(image, markers)` performs a marker-based image segmentation using the watershed algorithm.
    - Fernand Meyer. Color image segmentation. In Image Processing and its Applications, 1992., International Conference on, pages 303â€“306. IET, 1992.
    
**markers_updated = cv2.watershed(image, markers)**

Before passing the image to the function, you have to roughly outline the desired regions in the image markers with positive (>0) indices. So, every region is represented as one or more connected components with the pixel values 1, 2, 3, and so on. Such markers can be retrieved from a binary mask using findContours and drawContours (see the watershed.cpp demo). The markers are "seeds" of the future image regions. All the other pixels in markers , whose relation to the outlined regions is not known and should be defined by the algorithm, should be set to 0's. In the function output, each pixel in markers is set to a value of the "seed" components or to -1 at boundaries between the regions.

**Theory**

Any grayscale image can be viewed as a topographic surface where high intensity denotes peaks and hills while low intensity denotes valleys. You start filling every isolated valleys (local minima) with different colored water (labels). As the water rises, depending on the peaks (gradients) nearby, water from different valleys, obviously with different colors will start to merge. To avoid that, you build barriers in the locations where water merges. You continue the work of filling water and building barriers until all the peaks are under water. Then the barriers you created gives you the segmentation result. This is the "philosophy" behind the watershed.

* But from experiments, `cv2.watershed(image, markers)` asks image to be of RGB or BGR and markers to be of 32bit inteters

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
print (cv2.__version__)
```

```python
src_rgb = cv2.imread ('water_coins.jpg')
src_rgb = src_rgb.transpose(1,0,2)
im = cv2.cvtColor (src_rgb, cv2.COLOR_BGR2GRAY)
im.shape
```

```python
def cimshow(im, title=None, figsize=(10,10)):
    plt.figure(figsize=figsize)
    plt.imshow(im)
    plt.xticks([])
    #plt.yticks([])
    if title:
        plt.title(title)
        
def gimshow(im, title=None):
    plt.imshow(im, cmap='gray')
    plt.xticks([])
    #plt.yticks([])
    if title:
        plt.title(title)
gimshow(im)
```

## Thresholding

```python
ret, thresholded = cv2.threshold(im, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
print ('threshold: ', ret)
gimshow (thresholded)
```

## Some binary noise removal by Opening

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
im_opened = cv2.morphologyEx (thresholded, cv2.MORPH_OPEN, kernel, iterations=3)
gimshow(im_opened)
```

## Sure background

```python
sure_bg = cv2.dilate (im_opened, kernel, iterations=3)
gimshow(sure_bg, 'Black Area is Sure BG')
```

## Sure foreground
1. Distance Transform: Calculates the distance to the closest zero pixel for each pixel of the source image.
    - https://docs.opencv.org/2.4/modules/imgproc/doc/miscellaneous_transformations.html#distancetransform
    - It produces a gray scale image with `float32` data type

```python
dist_transform = cv2.distanceTransform(im_opened, cv2.DIST_L2, maskSize=5)
print (dist_transform.dtype)
gimshow(dist_transform, 'Distance Transformed')
```

```python
_ = plt.hist (dist_transform.ravel())
plt.title ('Histogram of distance transformed')
```

Sure foreground area is chosen by thresholding. This time the threshold is pre-determined as 0.7 of max of the distance transformed image. This is possible because the user assumes that the shape of the objects are almsot the same and therefore each object (coin) will have similar max distance to the others.

```python
ret, sure_fg = cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)
sure_fg = np.uint8 (sure_fg)
print (sure_fg.dtype)
gimshow (sure_fg, 'White Area is Sure Foreground')
```

## Unknown Region
The other areas all belong to unknown region. The subtraction results in a binary image too since `sure_bg` has wider white area.

```python
unknown = cv2.subtract (sure_bg, sure_fg)
gimshow(unknown, 'White Area is Unknown')
```

```python
np.unique (unknown)
```

## Connected Component Labeling
It labels background of the image with 0, then other objects are labelled with integers starting from 1.

```python
ret, im_ccl = cv2.connectedComponents(sure_fg)
print(im_ccl.dtype)
#
plt.imshow (im_ccl)
plt.colorbar()
plt.xticks([])
```

```python
# now, add 1 to im_ccl, any non-zero pixel is regarded as known by cv2.watershed()
markers = im_ccl + 1

# unknown area must have 0 values for watershed algorithm
markers[unknown==255] = 0
```

```python
# BGR image input, im_ccl: int32
markers = cv2.watershed(src_rgb, markers)
```

```python
plt.imshow (markers)
```

```python
seg_boundarys = src_rgb.copy()
seg_boundarys[markers == -1] = [255, 0, 0]

plt.figure (figsize=(10,10))
plt.imshow (seg_boundarys)
```

Note: `markers` image has -1 for the boundary pixels, 1 for the background, and 2 and above for the other object blobs.

```python
np.unique (markers)
```

```python
def segWatershed(src):
    im = cv2.cvtColor (src, cv2.COLOR_BGR2GRAY)
    ret, thresholded = cv2.threshold(im, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
    im_opened = cv2.morphologyEx (thresholded, cv2.MORPH_OPEN, kernel, iterations=3)
    sure_bg = cv2.dilate (im_opened, kernel, iterations=3)
    dist_transform = cv2.distanceTransform(im_opened, cv2.DIST_L2, maskSize=5)
    ret, sure_fg = cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)
    sure_fg = np.uint8 (sure_fg)
    unknown = cv2.subtract (sure_bg, sure_fg)
    ret, im_ccl = cv2.connectedComponents(sure_fg)
    # now, add 1 to im_ccl, any non-zero pixel is regarded as known by cv2.watershed()
    markers = im_ccl + 1

    # unknown area must have 0 values for watershed algorithm
    markers[unknown==255] = 0

    # Watershed
    markers = cv2.watershed(src_rgb, markers)
    
    return markers
```

```python
src_med = cv2.medianBlur (src_rgb, 5)
markers2 = segWatershed (src_med)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(markers2)
m8 = np.uint8(markers2)
cv2.imwrite ('markers.png', m8)
```

## Post Processing

To use `findCountours()` or `connectedComponents()`, let us set pixels of 1 and -1 to 0. Actually, all the information can be computed directly from `markers` image but instead of making our own program codes, we want to use ready-made functions.

```python
def getBinary(m):
    m = m.copy() - 1
    m[m <= 0] = 0
    print (m.dtype, binary.max(), binary.min(), np.unique(binary))
    return np.uint8(m)
```
