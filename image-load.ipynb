# Image Load/Save

- Various ways to load image files in jpg, png, etc

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
```

```python
imgurl = 'https://upload.wikimedia.org/wikipedia/commons/3/33/HSV_color_solid_cylinder_saturation_gray.png'
filename = 'saved_file.jpg'
```

## imageio
- https://imageio.readthedocs.io/en/stable/index.html

```python
import imageio

imio = imageio.imread (imgurl)

print ('imageio.imread(): ', type(imio), ' a subclass of np.ndarray')
#print ('meta: ', imio.meta)
plt.imshow (imio)
plt.title ('imageio RGB')
plt.pause(2)
```

```python
# pure numpy array
ii = np.asarray (imio)

plt.imshow (ii)
```

### File Save

```python
imageio.imsave(filename, imio)
print ('file saved to: ', filename)
```

```python
import os
os.listdir ('./')
```

## OpenCV

```python
import cv2 as cv
```

```python
imcv = cv.imread(filename)

print ('opencv.imread(): ', type(imcv))
```

```python
plt.imshow (imcv)
plt.title ('opencv BGR')
plt.show()
```

## PIL (Pillow Python Image Library)

```python
# using PIL
from PIL import Image

impil = Image.open (filename)

print ('PIL.Image.open(): ', type(impil))
```

```python
plt.imshow (impil)
plt.title ('PIL.Image RGB')
plt.pause(2)
```
